# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
dataset = pd.read_csv('heart11.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder() 
le2 = LabelEncoder() 
le6 = LabelEncoder() 
le8 = LabelEncoder() 
le10 = LabelEncoder() 
x[:,1] = le1.fit_transform(x[:,1])
x[:,2] = le2.fit_transform(x[:,2])
x[:,6] = le6.fit_transform(x[:,6])
x[:,8] = le8.fit_transform(x[:,8])
x[:,10] = le10.fit_transform(x[:,10])
print(dataset.shape)
print(dataset.dtypes)
```

```python
from pandas.plotting import scatter_matrix
import seaborn as sns

plt.figure(figsize=(10,10))
sns.heatmap(dataset.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
age_unique=sorted(dataset.Age.unique())
age_thalach_values=dataset.groupby('Age')['Oldpeak'].count().values
mean_thalach=[]
for i,Age in enumerate(age_unique):
    mean_thalach.append(sum(dataset[dataset['Age']==Age].Oldpeak)/age_thalach_values[i])
    
plt.figure(figsize=(10,5))
sns.pointplot(x=age_unique,y=mean_thalach,color='red',alpha=0.8)
plt.xlabel('Age',fontsize = 15,color='blue')
plt.xticks(rotation=45)
plt.ylabel('Oldpeak',fontsize = 15,color='blue')
plt.title('Age vs Oldpeak',fontsize = 15,color='blue')
plt.grid()
plt.show()
```

# Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X = dataset.drop(["HeartDisease"],axis=1)
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
from adjustText import adjust_text

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_test)
```

```python
print(X_train)
```

# Training Dataset

## Training with Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_scores = []
estimators = [10, 100, 200, 500, 1000]
for i in estimators:
    model_randomforest = RandomForestClassifier(n_estimators = i, random_state = 0)
    model_randomforest.fit(X_train, Y_train)
    rf_scores.append(model_randomforest.score(X_test, Y_test))
```

```python
from matplotlib.cm import rainbow

colors = rainbow(np.linspace(0, 1, len(estimators)))
plt.bar([i for i in range(len(estimators))], rf_scores, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores[i], "{:.2f}".format(rf_scores[i]))
    
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators')
```

```python
model_randomforest = RandomForestClassifier(n_estimators = 500)
model_randomforest.fit(X_train,Y_train)
```

## Training with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression(solver='lbfgs',max_iter=1000)
model_logistic.fit(X_train,Y_train)
```

## Training with KNeighbours Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn_scores = []
for k in range(1,21):
    model_kneighbors = KNeighborsClassifier(n_neighbors = k)
    model_kneighbors.fit(X_train, y_train)
    knn_scores.append(model_kneighbors.score(X_test, y_test))
```

```python
from matplotlib.ticker import FormatStrFormatter


plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    texts = [plt.text(i, knn_scores[i-1], (i, "{:.2f}".format(knn_scores[i-1])),fontsize='x-small')]
    adjust_text(texts,arrowprops=dict(arrowstyle='->', color='red'))

plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
model_kneighbors = KNeighborsClassifier(n_neighbors = 9)
model_kneighbors.fit(X_train,Y_train)
```

## Training with Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier


dt_scores = []
for i in range(1, len(X.columns) + 1):
    model_decision = DecisionTreeClassifier(max_features = i, random_state = 0)
    model_decision.fit(X_train, y_train)
    dt_scores.append(model_decision.score(X_test, y_test))
```

```python
plt.plot([i for i in range(1, len(X.columns) + 1)], dt_scores, color = 'green')
for i in range(1, len(X.columns) + 1):
    plt.text(i, dt_scores[i-1], (i, "{:.2f}".format(dt_scores[i-1])))
plt.xticks([i for i in range(1, len(X.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features')
```

```python
model_decision = DecisionTreeClassifier(max_features = 8)
model_decision.fit(X_train,Y_train)
```

## Training with XGBoost Classifier

```python
from xgboost import XGBClassifier
model_xgb = XGBClassifier(use_label_encoder=False)
model_xgb.fit(X_train,Y_train, eval_metric='rmse')
```

## Training with SVC

```python
from sklearn.svm import SVC

svc_scores = []
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for i in range(len(kernels)):
    model_svm = SVC(kernel = kernels[i])
    model_svm.fit(X_train, Y_train)
    svc_scores.append(model_svm.score(X_test, Y_test))
```

```python
colors = rainbow(np.linspace(0, 1, len(kernels)))
plt.bar(kernels, svc_scores, color = colors)
for i in range(len(kernels)):
    plt.text(i, svc_scores[i], "{:.2f}".format(svc_scores[i]))
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.title('Support Vector Classifier scores for different kernels')
```

```python
model_svm = SVC(kernel = 'poly')
model_svm.fit(X_train,Y_train)
```

# Making Confusion Matrix

```python
from sklearn.metrics import confusion_matrix , accuracy_score
# cm = confusion_matrix(Y_test,y_pred_random)
# print(cm)
```

```python
y_pred_logistic = model_logistic.predict(X_test)
y_pred_neighbors = model_kneighbors.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_decision = model_decision.predict(X_test)
y_pred_random = model_randomforest.predict(X_test)
y_pred_xgb = model_xgb.predict(X_test)
```

```python
RandomForest_Accuracy = accuracy_score(Y_test,y_pred_random)
LogisticRegression_Accuracy = accuracy_score(Y_test,y_pred_logistic)
KNeighbors_Accuracy = accuracy_score(Y_test,y_pred_neighbors)
SVM_Accuracy = accuracy_score(Y_test,y_pred_svm)
Decision_Accuracy = accuracy_score(Y_test,y_pred_decision)
XGBoost_Accuracy = accuracy_score(Y_test,y_pred_xgb)

print(RandomForest_Accuracy)
print(LogisticRegression_Accuracy)
print(KNeighbors_Accuracy)
print(SVM_Accuracy)
print(Decision_Accuracy)
print(XGBoost_Accuracy)
```
